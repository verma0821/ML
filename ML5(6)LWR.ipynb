```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("BostonHousing.csv")

# Take only one feature
X = df['rm'].values
Y = df['medv'].values

# Sort data for smooth graph
sort_idx = np.argsort(X)
X = X[sort_idx]
Y = Y[sort_idx]

# Bandwidth parameter
tau = 0.5

Y_pred = []

# Locally Weighted Regression
for x0 in X:

    # Gaussian weights
    w = np.exp(-(X - x0)**2 / (2 * tau**2))

    # Add bias term
    X1 = np.c_[np.ones(len(X)), X]

    # Diagonal weight matrix
    W = np.diag(w)

    # Calculate theta
    theta = np.linalg.pinv(X1.T @ W @ X1) @ (X1.T @ W @ Y)

    # Prediction
    y0 = theta[0] + theta[1] * x0

    Y_pred.append(y0)

# Plot graph
plt.scatter(X, Y, color='blue', label='Data')
plt.plot(X, Y_pred, color='red', label='LWR Curve')

plt.xlabel("Rooms (RM)")
plt.ylabel("Price (MEDV)")
plt.legend()

plt.show()
```
