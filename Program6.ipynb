```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("BostonHousing.csv")

# Take only one feature
X = df['rm'].values
Y = df['medv'].values

# Sort data for smooth curve
sort_idx = np.argsort(X)
X = X[sort_idx]
Y = Y[sort_idx]

# Bandwidth parameter
tau = 0.5

# Store predictions
Y_pred = []

# LWR Algorithm
for x0 in X:
    # Compute weights using Gaussian kernel
    w = np.exp(-(X - x0)**2 / (2 * tau**2))
    
    # Design matrix
    X1 = np.c_[np.ones(len(X)), X]
    
    # Weight matrix
    W = np.diag(w)
    
    # Normal equation (theta calculation)
    theta = np.linalg.pinv(X1.T @ W @ X1) @ (X1.T @ W @ Y)
    
    # Prediction
    y0 = theta[0] + theta[1] * x0
    Y_pred.append(y0)

# Plot graph
plt.scatter(X, Y, color='blue', label='Data')
plt.plot(X, Y_pred, color='red', label='LWR Curve')
plt.xlabel("Rooms (RM)")
plt.ylabel("Price (MEDV)")
plt.legend()
plt.title("Locally Weighted Regression")
plt.show()
```
